```python
win = 50
soltab  = [116  , 116  , 114,   95 ]
minttab = [20.10, 18.94, 20.00, 21.04]
maxttab = [20.30, 19.04, 20.16, 21.16]
for sol in soltab:
    print sol, apss_lib.insight_sol2ls(sol)
```
