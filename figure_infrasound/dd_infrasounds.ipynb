```python
import apss_lib
import numpy as np
import wavelet
%matplotlib inline
```

```python
mint = 14.588
maxt = 14.6
win = 2
```

```python
for sol in [26]:
    data = apss_lib.getsol(sol,var="PRE",distant=True,event=True)
    zefreq = 20.
    apss_lib.plotvar(data,code=["P_20Hz"],timetype="LMST",mint=mint,maxt=maxt,\
                     filename="dd_infrasound_sol%i"%(sol),fmt="%.1f")
    field,foo = apss_lib.plotvar(data,code=["P_20Hz"],\
                                 timetype="LMST",mint=mint,maxt=maxt,\
                                 window=win,fmt="%.1f",perturbamp=0.4,
                                 filename="dd_infrasound_sol%i"%(sol))
    wavelet.wavelet(field,freq=zefreq,ymin=0.2,ymax=10.,filename="dd_infrasound_sol%i"%(sol),s0fac=2.)
```
