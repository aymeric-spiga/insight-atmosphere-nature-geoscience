```python
import apss_lib
import numpy as np
import wavelet
%matplotlib inline
```

```python
mint = 21.75
maxt = 23.00
win = 500
```

```python
for sol in [78,75]:
    data = apss_lib.getsol(sol,var="PRE",distant=True)
    zefreq = np.min(data["PRESSURE_FREQUENCY"])
    apss_lib.plotvar(data,code=["PRE"],timetype="LTST",mint=mint,maxt=maxt,filename="gw_infrasound_sol%i"%(sol))
    field,foo = apss_lib.plotvar(data,code=["PRE"],timetype="LTST",\
                                 mint=mint,maxt=maxt,window=win,fmt="%.3f",perturbamp=0.125,
                                 filename="gw_infrasound_sol%i"%(sol))
    wavelet.wavelet(field,freq=zefreq,ymin=20,ymax=2000.,cutoff=100.,filename="gw_infrasound_sol%i"%(sol))
```
